Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

ans :- 

DataFrame:
    A DataFrame is a two-dimensional, labeled data structure in Pandas. It resembles a table with rows and columns.

    Key characteristics:
        Contains potentially different types of data in its columns.
        Supports both row and column labels.
        Can be thought of as a dictionary-like container for Series objects.
    Use cases:
    
        Handling structured data, such as CSV files, Excel spreadsheets, or database tables.
        Manipulating and analyzing tabular data.
        
Example of creating a DataFrame from a dictionary:

```python
#Example:-
data = {'Name' : ['Alice' , '25' , 'Female'],
       'Age' : ['Bob' , '30' , 'Male'],
       'Gender':['Claire' , '27', 'Female']}
pd.DataFrame(data)
```

Series:

    A Series is a one-dimensional array-like data structure in Pandas. It represents a single column of data.
    
    Key characteristics:
        Contains homogeneous data (i.e., data of the same type) in a single column.
        Has an associated index (usually numeric or labeled).
        
    Use cases:
        Representing a single feature or variable.
        Extracting specific columns from a DataFrame.
        
        
Example of creating a Series from a list:

```python
#Example:- 
data = [4,8,15,16,23,42]
my_series = pd.Series(data)
print(my_series)
```

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

    Among the given options, DataFrame is mutable in nature. A DataFrame allows modifications to its contents, such as adding or removing columns, changing values, and reordering rows. On the other hand, a Series (which represents a single column) is also mutable, but a Panel (which is a three-dimensional data structure) has been deprecated in recent versions of Pandas and is no longer commonly used. Therefore, the correct answer is DataFrame.
